```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xls = pd.ExcelFile("scale.xlsx")
sheetX = xls.parse(0)
```

```python
df = sheetX.dropna()
df
```

```python
df = df.rename(columns={'Unnamed: 0': 'nsteps', 'Unnamed: 1': 'python','Unnamed: 2': 'numba','Unnamed: 3': 'raw_cython','Unnamed: 4': 'cdef','Unnamed: 5': 'cmath','Unnamed: 6': 'numpy', 'Unnamed: 7': 'numba_omp', 'Unnamed: 8': 'cython_cmath'})
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
for column in df.columns[1:]:
    if column != 'x':  # Skip 'x' column if it's not numerical
        plt.plot(df['scale stable 50'], df[column], label=column)

plt.xlabel('scale')
plt.ylabel('time in seconds')
plt.title('Plot of all methods')
plt.legend()
plt.grid(True)
plt.show()
```

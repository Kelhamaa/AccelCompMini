```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xls = pd.ExcelFile("AccelComp.xlsx")
sheetX = xls.parse(0)
```

```python
sheetX.set_index("Unnamed: 0")
df = sheetX.dropna()
```

```python
df = df.rename(columns={'Unnamed: 0': 'nsteps', 'Unnamed: 1': 'python','Unnamed: 2': 'numba','Unnamed: 3': 'raw_cython','Unnamed: 4': 'cdef','Unnamed: 5': 'cmath','Unnamed: 6': 'numpy',})
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
for column in df.columns[1:]:
    if column != 'x':  # Skip 'x' column if it's not numerical
        plt.plot(df['nsteps'], df[column], label=column)

plt.xlabel('x')
plt.ylabel('Values')
plt.title('Plot of all columns')
plt.legend()
plt.grid(True)
plt.show()
```
